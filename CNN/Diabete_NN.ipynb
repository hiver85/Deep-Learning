```python
# load pima indians dataset
import pandas as pd 
ds = pd.read_csv("//content/drive/My Drive/Colab Notebooks/pima-indians-diabetes.data.csv", delimiter=",")
```

```python
ds.head(3)
```

```python
Y=ds.iloc[:,-1]
X=ds.drop(ds.columns[-1],axis=1)
```

```python
X.head(3)
```

```python
# create model
from keras.layers.advanced_activations import LeakyReLU
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu')),
model.add(Dense(1, activation='sigmoid'))
```

```python
# create model 2
from keras.layers.advanced_activations import LeakyReLU
model2 = Sequential()
model2.add(Dense(12, input_dim=8, activation='relu'))
model2.add(LeakyReLU(alpha=0.1))
model2.add(Dense(8, activation='relu')),
model2.add(LeakyReLU(alpha=0.1))
model2.add(Dense(1, activation='sigmoid'))
```

```python
# create model 3
from keras.layers.advanced_activations import LeakyReLU
model3 = Sequential()
model3.add(Dense(36, input_dim=8, activation='relu'))
model3.add(Dense(12, activation='relu')),
model3.add(Dense(8, activation='relu')),
model3.add(Dense(1, activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model2.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model3.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33)

epochs = 150
```

```python

# Fit the model
m1=model.fit(X_train, y_train, epochs=epochs, batch_size=10,verbose=1,validation_data=(X_test, y_test))
```

```python
# Fit the model
m2=model2.fit(X_train, y_train, epochs=epochs, batch_size=10,verbose=1,validation_data=(X_test, y_test))
```

```python
# Fit the model
m3=model3.fit(X_train, y_train, epochs=epochs, batch_size=10, verbose=1, validation_data=(X_test, y_test))
```

```python
predictions = model.predict(X_test)

# round predictions
rounded = [round(x[0]) for x in predictions]
print(rounded)
```

```python
score, acc = model.evaluate(X_test,y_test,batch_size=10)
print("test score: ", score)
print("test acc: ", acc)
```

```python
score, acc = model2.evaluate(X_test,y_test,batch_size=10)
print("test score: ", score)
print("test acc: ", acc)
```

```python
score, acc = model3.evaluate(X_test,y_test,batch_size=10)
print("test score: ", score)
print("test acc: ", acc)
#==> Overfit
```

```python
accuracy1 = m1.history['acc']
val1 = m1.history['val_acc']
loss1 = m1.history['loss']
val_loss1 = m1.history['val_loss']

accuracy2 = m2.history['acc']
val2 = m2.history['val_acc']
loss2 = m2.history['loss']
val_loss2 = m2.history['val_loss']

accuracy3 = m3.history['acc']
val3 = m3.history['val_acc']
loss3 = m3.history['loss']
val_loss3 = m3.history['val_loss']
```
